# **Importa biblioteca e arquivo Json**

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://github.com/alura-cursos/python-analise-chatgpt-assistente/raw/main/Dados/dados_vendas.json'
```

```python
# Lendo arquivo em formato json
df = pd.read_json(url)
```

```python
df.head()
```

```python
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# Normalizando as colunas item e loja que estão em formato dicionario oriundas de um arquivo json.

df = df.join(pd.json_normalize(df.item))
df = df.join(pd.json_normalize(df.loja))
df.head()
```

```python
# Dropa colunas aninhadas em dicionario após serem normalizadas

df.drop(columns=['item', 'loja'], inplace=True)
df.head()
```

# **Tratando dados nulos no dataframe**

```python
## Verifica se existem valores nulos no Dataframe
df.isnull().sum()
```

```python
## Trata dados nulos do dataframe

df['item_quantidade_venda'].fillna(0, inplace=True)
df['item_peso'].fillna(df['item_peso'].mean(), inplace=True)
df['loja_tamanho'].fillna('Não Informado', inplace=True)
df.head()
```

```python
## Verifica se ainda existem valores nulos no Dataframe
df.isnull().sum()
```

# **Padroniza dados categoricos**

```python
## verifica valores unicos na coluna, para padronização
df['item_conteudo_gordura'].unique()
```

```python

# Aplicar o mapeamento usando o método loc
df.loc[df['item_conteudo_gordura'].isin(['reg']), 'item_conteudo_gordura'] = 'Regular'
df.loc[df['item_conteudo_gordura'].isin(['BTG', 'baixo teor de gordura']), 'item_conteudo_gordura'] = 'Baixo Teor de Gordura'

# Verificar os valores únicos na coluna após o mapeamento
df['item_conteudo_gordura'].unique()
```
